```python
import time
```

### Unoptimized code 1

I wrap it in a for loop to run it a few times, so we can see the difference

```python
# This code calculates the sum of all numbers in the list that are multiples of 10
t = time.time()
for i in range(1000):
    numbers = [10, 20, 30, 40, 50]
    total = 0
    for num in numbers:
        if num % 10 == 0:
            total += num
    print('Total sum of multiples of 10 is:', total)
time.time()-t
```

### Optimized versions

```python
t = time.time()
for i in range(1000):
    numbers = [10, 20, 30, 40, 50]
    print(sum(num for num in numbers if num % 10 == 0))
time.time()-t
```

```python
from functools import reduce
t = time.time()
for i in range(1000):
    print(reduce(lambda x, y: x + y, filter(lambda x: x % 10 == 0, [10, 20, 30, 40, 50])))
time.time()-t
```

### Unoptimized code 2

```python
t = time.time()
for i in range(1000):
    sentence = 'The quick brown fox jumps over the lazy dog'
    long_words = []
    for word in sentence.split():
        if len(word) > 4:
            long_words.append(word)
    print('Words longer than 4 characters:', long_words)
time.time()-t
```

### Optimized version

```python
t = time.time()
for i in range(1000):
    sentence = 'The quick brown fox jumps over the lazy dog'
    print('Words longer than 4 characters:', [word for word in sentence.split() if len(word) > 4])
time.time()-t
```

```python
import re
t = time.time()
for i in range(1000):
    sentence = 'The quick brown fox jumps over the lazy dog'
    print('Words longer than 4 characters:', [word for word in re.findall(r'\w+', sentence) if len(word) > 4])
time.time()-t
```

```python
from pandas import Series
t = time.time()
for i in range(1000):
    sentence = 'The quick brown fox jumps over the lazy dog'
    print('Words longer than 4 characters:', Series(sentence.split()).str.len()[Series(sentence.split()).str.len() > 4].tolist())
time.time()-t
```

### Unoptimized code 3

```python
t = time.time()
for i in range(1000):
    text = 'programming is fun'
    repeated_chars = []
    for char in text:
        if text.count(char) > 1 and char not in repeated_chars:
            repeated_chars.append(char)
    print('Characters that appear more than once:', repeated_chars)
time.time()-t
```

### Optimized version

```python
from collections import Counter
t = time.time()
for i in range(1000):
    text = 'programming is fun'
    print('Characters that appear more than once:', Counter(text).most_common(1)[0][0])
time.time()-t
```

```python
t = time.time()
for i in range(1000):
    print('Characters that appear more than once:', [char for char in 'programming is fun' if 'programming is fun'.count(char) > 1])
time.time()-t
```

```python
t = time.time()
for i in range(1000):
    print('Characters that appear more than once:', {char: 'programming is fun'.count(char) for char in 'programming is fun' if 'programming is fun'.count(char) > 1}.keys())
time.time()-t
```
